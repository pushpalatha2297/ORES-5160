```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('brain_stroke.csv')
```

#Bar Chart
A bar chart is a graphical representation of data that uses bars to compare different categories of data. Each bar represents a category of data, and the length or height of the bar corresponds to the value or frequency of that category.


```python
stroke_counts = df.groupby('gender')['stroke'].sum()

# Plot the bar chart
stroke_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Stroke Incidence by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Stroke Incidents')
plt.xticks(rotation=0)  # Keep the gender labels horizontal
plt.show()
```

#Line Chart
A line chart is a type of chart that displays information as a series of data points, known as 'markers,' connected by straight line segments.

```python
stroke_patients = df[df['stroke'] == 1]

# Sort the DataFrame by age for plotting
sorted_patients = stroke_patients.sort_values(by='age')

# Plotting the line graph
plt.plot(sorted_patients['age'], sorted_patients['avg_glucose_level'], marker='o')
plt.title('Average Glucose Level by Age for Stroke Patients')
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.grid(True)
plt.show()
```

#Pie Chart
The pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. Each slice of the pie represents a category and its size is proportional to the frequency or quantity it represents in the whole.

```python
work_type_counts = df['work_type'].value_counts()

# Plot the pie chart
plt.pie(work_type_counts, labels=work_type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Work Types Among Patients')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

#Histogram
A histogram is a type of graph that represents the distribution of numerical data by way of bins. Its similar to a bar chart but with a focus on ranges or intervals of data, known as 'bins,' rather than individual categories.

```python
plt.hist(df[df['stroke'] == 1]['age'], bins=10, color='green', alpha=0.7)
plt.title('Age Distribution of Stroke Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#Box Plot
Boxplots encapsulate the distribution of data into five key characteristics  the minimum, first quartile, median, third quartile, and maximum  offering a quick visual overview of the central tendency, spread, and skewness of the data, as well as identifying outliers. They are invaluable for comparing distributions across different categories.

```python
sns.boxplot(x='stroke', y='bmi', data=df)
plt.title('BMI Distribution by Stroke Status')
plt.xlabel('Stroke Status (0: No, 1: Yes)')
plt.ylabel('BMI')
plt.show()
```

#Scatter Plot
Scatter plots present the relationship between two quantitative variables, offering insights into correlation trends, data clustering, and outlier points. Each point on the plot corresponds to one data observation, making it straightforward to perceive the distribution and direction of data.

```python
sns.scatterplot(x='avg_glucose_level', y='bmi', hue='stroke', data=df)
plt.title('Glucose Level vs. BMI for Stroke Patients')
plt.xlabel('Average Glucose Level')
plt.ylabel('BMI')
plt.legend(title='Stroke Status', labels=['No Stroke', 'Stroke'])
plt.show()
```
